```python
import requests

from bs4 import BeautifulSoup
import pandas as pd

# URL of the website to scrape
url = "https://www.imdb.com/chart/top"

# nHTTP GET request to the website
dff = requests.get(url)

# Parse the HTML code using BeautifulSoup
soup = BeautifulSoup(dff.content, 'html.parser')

# Extract the relevant information from the HTML code
movies = []
for row in soup.select('tbody.lister-list tr'):
  tle = row.find('td', class_='titleColumn').find('a').get_text()
  year = row.find('td', class_='titleColumn').find('span', class_='secondaryInfo').get_text()[1:-1]
  movies.append([tle, year])

df = pd.DataFrame(movies)
df.to_csv('data_mobile.csv', index=False)
```

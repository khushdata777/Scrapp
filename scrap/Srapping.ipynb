```python
import requests
from bs4 import BeautifulSoup
response = requests.get("https://en.wikipedia.org/wiki/Mughal_dynasty#:~:text=The%20Mughal%20empire%20is%20conventionally,moved%20towards%20the%20Indian%20subcontinent.")

soup = BeautifulSoup(response.content, "html.parser")

text = soup.get_text()
print(text)
```
